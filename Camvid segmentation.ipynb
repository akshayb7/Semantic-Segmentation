```python
codes = np.loadtxt(path/'codes.txt', dtype=str)
codes
```

### Dataset creation

```python
size = src_size//2
bs = 4
```

```python
src = (SegmentationItemList
       .from_folder(path_img)
       .split_by_fname_file('../valid.txt')
       .label_from_func(get_label, classes=codes))
src
```

```python
def get_data(SZ, BS):
    data = (src.transform(get_transforms(p_affine=0.8, max_lighting=0.3), size=SZ, tfm_y=True)
            .databunch(bs=BS)
            .normalize(imagenet_stats))
    return data
```

```python
data = get_data(size, bs)
```

```python
data.show_batch(2, figsize=(10,8))
```

```python
data.show_batch(2, figsize=(10,8), ds_type=DatasetType.Valid)
```

### Model creation

```python
name_to_id = {v:k for k,v in enumerate(codes)}
void_code = name_to_id['Void']
```

```python
def accuracy_camvid(input, target):
    target = target.squeeze(1)
    mask = target != void_code
    return (input.argmax(dim=1)[mask]==target[mask]).float().mean()
```

```python
metrics = accuracy_camvid
wd = 1e-2
```

```python
learn = unet_learner(data, models.resnet34, metrics=metrics, wd=wd)
```

```python
learn.lr_find()
learn.recorder.plot()
```

```python
lr = 2e-4
```

```python
learn.fit_one_cycle(cyc_len=10, max_lr = slice(lr), pct_start=0.9)
```

```python
learn.save('stage-1-small')
```

```python
learn.fit_one_cycle(cyc_len=6, max_lr = slice(lr), pct_start=0.8)
```

```python
learn.save('stage-1-small')
```

```python
learn.load('stage-1-small')
```

```python
learn.show_results(rows=2, figsize=(10,8))
```

```python
learn.unfreeze()
```

```python
learn.lr_find()
learn.recorder.plot()
```

```python
lrs = slice(lr/200, lr/20)
```

```python
learn.fit_one_cycle(cyc_len=15, max_lr=lrs, pct_start=0.8)
```

```python
learn.save('stage-2-small')
```

```python
learn.show_results(rows=2, figsize=(10,8))
```
